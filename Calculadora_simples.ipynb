```python
## Módulo 5 - Lógica de Programação com Python
## Calculadora simples para operações aritméticas fundamentais.

# Lista de operações para escolha do usuário
operacoes = [
    "Adição (+)",
    "Subtração (-)",
    "Multiplicação (*)",
    "Divisão (/)",
]

# Interação inicial com o usuário
name = input("Olá, caro estudante! Como você se chama? ")
print(f"\nQue tal aprendermos mais sobre matemática, {name}?")

# Loop para funcionamento da calculadora
while True:
    continuar = input("\nPressione ENTER para começar um cálculo ou digite 'sair' para fechar a calculadora: ").lower()

    if continuar == "sair":
        print(f"Obrigado por usar a calculadora, {name}! Até a próxima.")
        break

    try:
# 1. Entrada dos números
        num1 = float(input("Insira o primeiro número: "))
        num2 = float(input("Insira o segundo número: "))

# 2. Exibição da lista de operações
        print("\nEscolha a operação:")
        for i, op in enumerate(operacoes):
            print(f"[{i + 1}] - {op}")

# Inicialização do resultado
        resultado = None

## Loop de validação para a escolha da operação ##
        while True:
            escolha = int(input("Digite o número da operação: "))

            if 1 <= escolha <= len(operacoes):
                break # Sai do loop de validação interna
            else:
                print("Opção inválida. Por favor, digite um número entre 1 e 4.")

# 3. Execução da operação
        if escolha == 1:
            operacao_nome = "adição"
            resultado = num1 + num2

        elif escolha == 2:
            operacao_nome = "subtração"
            resultado = num1 - num2

        elif escolha == 3:
            operacao_nome = "multiplicação"
            resultado = num1 * num2

        elif escolha == 4:
            operacao_nome = "divisão"
            if num2 != 0:
                resultado = num1 / num2
            else:
                print("Erro! Divisão por zero não é permitida.")
                # resultado permanece None, impedindo a impressão

# 4. Impressão do resultado
        if resultado is not None:
            print(f"O resultado da {operacao_nome} é {resultado}.")

    except ValueError:
# Captura erros de entrada de número ou escolha de opção (se for texto)
        print("\nEntrada inválida. Por favor, digite apenas números para os valores e para a opção.")

    except Exception as e:
# Captura qualquer outro erro inesperado
        print(f"\nOcorreu um erro inesperado: {e}.")

# Separador visual
    print("-" * 25)
```
